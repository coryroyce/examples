```python
import numpy as np

vecs = np.random.rand(100, 3)
```

```python
import plotly.graph_objects as go

contexts = go.Scatter3d(
    x=vecs[:,0],
    y=vecs[:,1],
    z=vecs[:,2],
    marker=dict(
        color='rgba(28,23,255,0.5)',
    ),
    line=dict(
        width=0,
        color='rgba(0,0,0,0)'
    ),
    name="points"
)

axes = go.Scatter3d(
    x=[0, 1, 0, 0, 0, 0, 0],
    y=[0, 0, 0, 1, 0, 0, 0],
    z=[0, 0, 0, 0, 0, 1, 0],
    marker=dict(
        size=0,
        color='rgba(0,0,0,0)',
        symbol='cross'
    ),
    line=dict(
        width=3,
        color='rgba(0,0,20,1)'
    ),
    name="axes"
)
go.Figure(data=[contexts, axes])
```

```python
norm_vecs = vecs / np.linalg.norm(vecs, axis=1, keepdims=True)
```

```python
contexts = go.Scatter3d(
    x=norm_vecs[:,0],
    y=norm_vecs[:,1],
    z=norm_vecs[:,2],
    marker=dict(
        color='rgba(28,23,255,0.5)',
    ),
    line=dict(
        width=0,
        color='rgba(0,0,0,0)'
    ),
    name="points"
)

axes = go.Scatter3d(
    x=[0, 1, 0, 0, 0, 0, 0],
    y=[0, 0, 0, 1, 0, 0, 0],
    z=[0, 0, 0, 0, 0, 1, 0],
    marker=dict(
        size=0,
        color='rgba(0,0,0,0)',
        symbol='cross'
    ),
    line=dict(
        width=3,
        color='rgba(0,0,20,1)'
    ),
    name="axes"
)
go.Figure(data=[contexts, axes])
```
